```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([23.02,19.07,22.57,17.38,12.97])
y=np.array([75.57,72.87,88.36,78.48,77.59])
z=["Ahmedabad","Mumbai","Kolkata","Hyderabad","Bengaluru"]
plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
for i,label in enumerate(z):
  plt.annotate(label,(x[i],y[i]))
plt.show()
```
